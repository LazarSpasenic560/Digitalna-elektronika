```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Audio
```

Ucitavanje potrebnih biblioteka.

```python
fs = 8000
t = np.linspace(0, 1, fs)
```

Podesavanje parametara za iscrtavanje. Linija koja se iscrtava je sacinjena od tacaka, a broj tacaka se zadaje preko fs vrednosti. Vrednost t sluzi da zada funkciju np.linspace kojom se oznacava da izmedju tacaka nema razmaka i da se one dodiruju.

```python
def synthesize(sampling_speed, wavetable, n_samples):
    samples = []
    current_sample = 0
    while len(samples) < n_samples:
        current_sample += sampling_speed
        current_sample = current_sample % wavetable.size
        samples.append(wavetable[current_sample])
        current_sample += 1
    return np.array(samples)
```

Funkcija synthesize sluzi da na osnovu nacrtanog uzorka generise odgovarajuci audio zvuk.

```python
#sinusoida
wavetable = np.sin(2 * np.pi * t)
```

Funkcija plot sluzi za crtanje slike koju smo zadali.

```python
sample1 = synthesize(220, wavetable, 2 * fs)
sample2 = synthesize(440, wavetable, 2 * fs)
```

Ovde se generise na kojoj ce se frekvenciji zvuk cuti.
